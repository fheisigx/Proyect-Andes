```python
import pandas as pd
import glob
import numpy as np
import os

path = "Peajes/"
```

```python
allFiles = glob.glob(path + "*.xl*")
```

```python
allFiles
```

```python
frame = pd.DataFrame()
list_ = []
i=0
for file_ in allFiles:
    base = pd.read_excel(file_,sheetname="Cuadro de Pagos",skiprows=7) #,index_col=1
    base = base[['Usuaria','Total']]
    base.columns = ['Empresa', allFiles[i][7:12]]
    base = base.set_index('Empresa')
    base = base.ix[1:]
    
    list_.append(base)
    i=i+1
    
frame = pd.concat(list_,axis=1)

frame.to_excel('Peajes.xlsx')
```
